```python

missing_counts = df.isnull().sum()

missing_percentage = (missing_counts / len(df)) * 100

missing_percentage_sorted = missing_percentage.sort_values(ascending=False)

highlight_variable = 'UNION'

plt.figure(figsize=(10, 8))
colors = ['skyblue' if x != highlight_variable else 'orange' for x in missing_percentage_sorted.index]
plt.barh(missing_percentage_sorted.index, missing_percentage_sorted, color=colors)
plt.xlabel('Percentage of Missing Values (%)')
plt.title('Percentage of Missing Values by Variable')
plt.gca().invert_yaxis()
plt.tight_layout()

filename = "highlighted_missing_values_chart.png"
plt.savefig(filename)

plt.show()

files.download(filename)
```

```python
correlations = {}
for column in df.columns:
    if column != 'LFSSTAT':
        rho, pval = stats.spearmanr(df['LFSSTAT'], df[column], nan_policy='omit')
        correlations[column] = rho

ranked_correlations = sorted(correlations.items(), key=lambda x: x[1], reverse=True)

for name, corr in ranked_correlations:
    print(f"{name}: {corr}")
```

```python
correlations = {}
for column in df.columns:
    if column != 'LFSSTAT':
        non_null_data = df[['LFSSTAT', column]].dropna()
        rho, pval = stats.pearsonr(non_null_data['LFSSTAT'], non_null_data[column])
        correlations[column] = rho

ranked_correlations = sorted(correlations.items(), key=lambda x: abs(x[1]), reverse=True)

for name, corr in ranked_correlations:
    print(f"{name}: {corr}")
```

```python
df.drop(columns=['REC_NUM', 'SURVYEAR', 'SURVMNTH', 'PROV',
       'AGE_6', 'MJH', 'EVERWORK', 'FTPTLAST',
       'COWMAIN', 'NOC_43', 'YABSENT',
       'WKSAWAY', 'PAYAWAY', 'UHRSMAIN', 'AHRSMAIN', 'FTPTMAIN', 'UTOTHRS',
       'ATOTHRS', 'HRSAWAY', 'YAWAY', 'PAIDOT', 'UNPAIDOT', 'XTRAHRS', 'WHYPT',
       'TENURE', 'PREVTEN', 'HRLYEARN', 'PERMTEMP', 'ESTSIZE',
       'FIRMSIZE', 'DURUNEMP', 'FLOWUNEM', 'UNEMFTPT', 'WHYLEFTO', 'WHYLEFTN',
       'DURJLESS', 'AVAILABL', 'LKPUBAG', 'LKEMPLOY', 'LKRELS', 'LKATADS',
       'LKANSADS', 'LKOTHERN', 'PRIORACT', 'YNOLOOK', 'TLOLOOK', 'SCHOOLN',
       'EFAMTYPE', 'AGYOWNK', 'FINALWT'], inplace = True)
```

```python
nominal_columns = ['CMA', 'SEX', 'NAICS_21', 'NOC_10']

for column in nominal_columns:
    df[column] = df[column].astype('category')
```

```python
df['LFSSTAT'] = pd.Categorical(df['LFSSTAT'], categories=[1,2,3,4], ordered=True)
df['AGE_12'] = pd.Categorical(df['AGE_12'], categories=list(range(1, 13)), ordered=True)
df['MARSTAT'] = pd.Categorical(df['MARSTAT'], categories=[1,2,3,4,5,6], ordered=True)
df['EDUC'] = pd.Categorical(df['EDUC'], categories=[0,1,2,3,4,5,6], ordered=True)
df['IMMIG'] = pd.Categorical(df['IMMIG'], categories=[1,2,3], ordered=True)
df['UNION'] = pd.Categorical(df['UNION'], categories=[1,2,3], ordered=True)
```

```python
df.info()
```

```python
df = pd.get_dummies(df, columns=['SEX', 'CMA', 'NAICS_21', 'NOC_10'])
```
