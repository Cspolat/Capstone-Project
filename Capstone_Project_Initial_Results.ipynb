# Feature Engineering


Binarizing `LFSSTAT` to create a balanced binary class attribute. Originally there were 4 categories: Employed at work '1'; Employed, absent from work '2'; Unemployed '3'; and Not in labour force '4'. There is an order from 1 to 4 that refers to the transition from being employed to not in labour force category. First two were recatogorized into category 1, and the rest in category 0 the 0 to one arrangement was done by considering the ordinal nature of this variable. 0 is not-employed, 1 is employed.

```python
# Combining categories 2 into 1(Employed), and 4 into 3
df['LFSSTAT'] = df['LFSSTAT'].replace({2: 1, 4: 3})
# Converting 3 into 0(Not-Employed)
df['LFSSTAT'] = df['LFSSTAT'].replace({3: 0})
```

# `LOGISTIC REGRESSION MODEL`






Below this, the models will be evaluated for Alternative 2, which includes SMOTE

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import make_pipeline
from sklearn.metrics import f1_score

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

logreg_pipeline = make_pipeline(smote, LogisticRegression(C=14.528246637516036, solver='saga', max_iter=1000, random_state=42)) # C value and solver was determined using the hyperparameter tuning above.

f1_scores = []
for train_idx, test_idx in cv.split(X, y):  #This for loop is to divide the datasetinto folds using the cross-validation method.
    X_train_fold, X_test_fold = X.iloc[train_idx], X.iloc[test_idx]
    y_train_fold, y_test_fold = y.iloc[train_idx], y.iloc[test_idx]

    logreg_pipeline.fit(X_train_fold, y_train_fold)

    y_pred_fold = logreg_pipeline.predict(X_test_fold)

    fold_f1_score = f1_score(y_test_fold, y_pred_fold, average='binary')
    f1_scores.append(fold_f1_score)


print(f1_scores)
```

The following coefficients quantifies the relationship between employment status and other independent variables. For example, if NOC_10_1.0 (can be 0 or 1 only) goes from 0 to 1, this exlains the 0.728 increase in the employment status which is either 0 or 1. This indicates that Management Occupations (NOC_10_1.0) has a critical role in the explanation of employment status which is about 73%.

Using `FRIEDMAN TEST` for hypothesis testing for our three models

```python
from scipy.stats import friedmanchisquare

# F1 scores from the 3 models across 10 folds
Logistic_Regression= [0.895193320904596, 0.8969481097709059, 0.8935241329129275, 0.8988945372387638, 0.8936756012632602, 0.892706645056726, 0.8957760155364947, 0.8946515397082658, 0.8979492975940577, 0.8915388983532083]
XGBoost = [0.8878144718511841, 0.8898381664339111, 0.8879558738783239, 0.8890897140979297, 0.8880842659644503, 0.8908569085690857, 0.8855049798337311, 0.8923505780109864, 0.8920910430653348, 0.8899435398085264]
Random_Forest = [0.8685473684210527, 0.8715356751747957, 0.8701659785997136, 0.873582528349433, 0.8681476476053721, 0.8695725345001682, 0.8691541448590241, 0.873371985547433, 0.8763913298183948, 0.869997472407111]


print(f"Mean F1 Score for Logistic Regression: {np.mean(Logistic_Regression):.4f}")
print(f"Mean F1 Score for XGBoost: {np.mean(XGBoost):.4f}")
print(f"Mean F1 Score for Random Forest: {np.mean(Random_Forest):.4f}")

friedman = friedmanchisquare(Logistic_Regression, XGBoost, Random_Forest)
print(friedman.statistic)
print(friedman.pvalue)

alpha = 0.05
if friedman.pvalue < alpha:
    print('There is a statistically significant difference between the models (reject H0)')
else:
    print('There is not a statistically significant difference between the models (fail to reject H0)')
```

This result suggests that the null hypothesis can be rejected as p value is less than 5 percent. It can be observed from the results that XGBoost has the highest F1 Scores among the other models.
